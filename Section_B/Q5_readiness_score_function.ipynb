```python
def readiness_score(attendance, diary_entries, task_count):
    score = attendance * 0.5 + diary_entries * 2 + task_count * 3

    if score > 100:
        score = 100
    elif score < 0:
        score = 0
    else:
        score = score

    return score

print("Score out of 100:", readiness_score(85, 5, 8))
```
