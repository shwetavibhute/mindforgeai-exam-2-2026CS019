a = 25
b = a
c = 25
names = ["Asha", "Riya"]
same_names = names
same_names.append("Neha")

Explination:

a = 25 :- a points to integer 25.
b = a :- b points to the same object as a.
c = 25 :- c also points to the same object.
names = ["Asha", "Riya"] :- A new list is created.
same_names = names :- Both variables point to the same list.
same_names.append("Neha") :- The list is updated, so both variables show the change.
