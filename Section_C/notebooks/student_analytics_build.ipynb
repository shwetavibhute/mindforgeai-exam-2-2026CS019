3.Create a function that calculates final readiness score from attendance, diary, task, assignment, and lab signals. 

```python
def readiness_score(attendance, diary, tasks, assignment, lab):
    score = (
        attendance * 0.30 +
        diary * 2 +
        tasks * 3 +
        assignment * 0.25 +
        lab * 0.25
    )

    if score > 100:
        score = 100

    return round(score, 2)

df["readiness_score"] = df.apply(
    lambda x: readiness_score(
        x["attendance_pct"],
        x["diary_entries"],
        x["tasks_completed"],
        x["assignment_score"],
        x["lab_score"]
    ),
    axis=1
)

df[["name","readiness_score"]]
```

### Explanation

A function is created to calculate the readiness score using attendance, diary entries, completed tasks, assignment score, and lab score.

```python
4.Use loops or apply logic to classify each student into Support, Developing, Strong, or Excellent. 
```

```python
performance = []

for score in df["readiness_score"]:
    if score < 45:
        performance.append("Support")
    elif score < 70:
        performance.append("Developing")
    elif score < 90:
        performance.append("Strong")
    else:
        performance.append("Excellent")

df["Performance_Band"] = performance

df[["name","readiness_score","Performance_Band"]]
```

### Explanation

This code classifies each student into Support, Developing, Strong, or Excellent based on the readiness score using a loop.

```python
5.Generate at least three analytics outputs: department average, top performer, and students needing support.
```

```python
print("-----Department Average Readiness Score-----")
print(df.groupby("department")["readiness_score"].mean())

print("\n-----Top Performer-----")
top = df.loc[df["readiness_score"].idxmax()]
print(top[["name","department","readiness_score"]])

print("\n-----Students Needing Support------")
print(df[df["Performance_Band"]=="Support"][["name","department","readiness_score"]])
```

### Explanation

This code generates three analytics:
- Department-wise average readiness score.
- Top-performing student.
- Students who need support.
